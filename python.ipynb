### multi join 

```python
Stu_course.head()
```

```python
multi_join=pd.merge(inner_join_py[['Top_Name','Crs_Name','Crs_Id']],Stu_course,on='Crs_Id')
multi_join
```

## create a student exam_Statue

```python
booleans=[]
for item in multi_join.Grade:
    if item >=80:
        booleans.append('Sucess')
    else:
        booleans.append('Falied')
```

```python
status_pd=pd.DataFrame(booleans,columns =['sucess_or_fail'])
```

```python
status_pd.head()
```

```python
result=pd.concat([multi_join,status_pd],axis=1)
```

```python
result
```

## insert DataFrame Into SQl Server as Table Database

```python
for index,row in result.iterrows():
    cursor.execute('INSERT INTO dbo.python_table([Top_Name],[Crs_Name],[Crs_Id],[St_Id],[Grade],[sucess_or_fail]) values (?,?,?,?,?,?)', 
                    row['Top_Name'], 
                    row['Crs_Name'], 
                    row['Crs_Id'],
                    row['St_Id'],
                    row['Grade'],
                    row['sucess_or_fail'])
    conn.commit()
cursor.close()
conn.close()
```
